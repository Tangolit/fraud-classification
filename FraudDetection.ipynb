```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, recall_score, make_scorer

model = LogisticRegression()
credit_card_data = pd.read_csv("creditcard.csv")
credit_card_data
X = credit_card_data.drop(columns= ["Time", "Amount", "Class"])
Y = credit_card_data["Class"]

def min_score(y_true, y_pred):
    precision_score1 = precision_score(y_true, y_pred)
    recall_score1 = recall_score(y_true, y_pred)
    return min(precision_score1, recall_score1)
    
grid = GridSearchCV (
    estimator = model,
    param_grid = {"class_weight" : [{0: 1, 1: v }for v in np.linspace(1,20,30)]
                 },
    scoring = {"precision_score" : make_scorer(precision_score),
               "recall_score": make_scorer(recall_score),
               "min_both" : make_scorer(min_score)
              },
    refit = "min_both",
    cv = 4, 
    n_jobs = -1,
    return_train_score = True
                  
            )
grid.fit(X,Y)
print(grid.predict(X).sum())
print(Y.sum())
pd.DataFrame(grid.cv_results_)
```
